## 1. What is a Workflow?

A workflow is a predefined sequence of steps used to complete a task.

Example:

```text
Receive Request
      ↓
Validate Data
      ↓
Process Request
      ↓
Send Notification
```

Each step is known in advance.

````python
workflow_steps = [
    "Receive Request",
    "Validate Data",
    "Process Request",
    "Send Notification"
]

for step in workflow_steps:
    print(f"Executing step: {step}")
````

## 2. Workflow as Functions

A workflow can also be represented as several functions executed in a fixed order.

This is similar to how many scripts, pipelines, and automation jobs are structured.

````python
def receive_request():
    print("Request received")


def validate_data():
    print("Data validated")


def process_request():
    print("Request processed")


def send_notification():
    print("Notification sent")


receive_request()
validate_data()
process_request()
send_notification()
````

## 3. Simple Automation Example

Automation means executing repetitive tasks with minimal human intervention.

In this example, we process multiple orders using the same predefined steps.

````python
orders = ["ORD-001", "ORD-002", "ORD-003"]

for order in orders:
    print(f"Processing {order}")
    print(f"Validating {order}")
    print(f"Sending notification for {order}")
    print("---")
````

## 4. Rule-Based Workflow

Workflows often use rules.

Rules are useful when the process is predictable and the possible cases are known in advance.

````python
ticket_priority = "high"

if ticket_priority == "critical":
    print("Assign immediately to incident team")
elif ticket_priority == "high":
    print("Assign to support engineer")
elif ticket_priority == "medium":
    print("Add to normal queue")
else:
    print("Add to backlog")
````

## 5. Workflow Limitation

The more rules we add, the harder the workflow becomes to maintain.

Imagine having:

- 10 ticket types
- 5 priorities
- 20 teams
- different escalation paths
- different approval rules

This is where static workflows can become complex.

````python
user_request = input("Enter request type: ")

if "weather" in user_request.lower():
    print("Workflow action: call weather API")
elif "email" in user_request.lower():
    print("Workflow action: send email")
elif "ticket" in user_request.lower():
    print("Workflow action: create ticket")
else:
    print("Workflow action: unknown request")
````

## 6. Workflow vs Agent

A workflow follows predefined rules.

An agent decides dynamically what to do based on:

- the user goal
- the available context
- the available tools
- the current state of the task

| Workflow | Agent |
|---|---|
| Fixed steps | Dynamic decisions |
| Rule-based | Goal-based |
| Predictable | Adaptive |
| Executes process | Reasons about process |

## 7. Pseudo-Agent Example

This is **not a real AI Agent yet**.

But it demonstrates the idea of selecting an action based on user intent.

A real agent would usually use an LLM to reason about the request and decide which tool to call.

````python
def choose_action(user_request):
    request = user_request.lower()

    if "weather" in request:
        return "Call Weather Tool"
    elif "email" in request:
        return "Call Email Tool"
    elif "summary" in request or "summarize" in request:
        return "Call Summarization Tool"
    elif "ticket" in request:
        return "Call Ticketing Tool"
    else:
        return "Ask for clarification"


user_request = input("What do you need? ")

selected_action = choose_action(user_request)

print(f"User request: {user_request}")
print(f"Selected action: {selected_action}")
````

## 8. Workflow + Agent

Modern systems often combine both:

```text
Workflow
    ↓
Agent
    ↓
Workflow
```

Example:

1. Workflow receives a request
2. Agent analyzes the request
3. Workflow executes the selected action

````python
def workflow_start(user_request):
    print("Workflow started")
    return user_request


def pseudo_agent_analyze(user_request):
    print("Agent analyzing request...")
    return choose_action(user_request)


def workflow_execute(action):
    print(f"Workflow executing: {action}")


request = workflow_start("Please summarize this document")
action = pseudo_agent_analyze(request)
workflow_execute(action)
````

## 10. Key Takeaways

- Workflows are predefined sequences of steps
- Automation reduces manual work
- Workflows are predictable and repeatable
- Workflows become difficult with complex decision making
- Agents can make dynamic decisions
- Workflows and agents can work together
