```python
# Session 1 - Introduction to AI Agents

In this notebook we will:

- Make our first LLM call
- Ask questions to an LLM
- Build a simple chat application
- Understand the difference between an LLM and an Agent
```

```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI()
```

## Step 2 - First LLM Call

Now we will send our first question to the LLM.

```python
response = client.responses.create(
    model="gpt-5",
    input="Explain what an AI Agent is in one short paragraph."
)

print(response.output_text)
```

## Discussion

Is this already an AI Agent?

No. This is only a direct call to an LLM.

At this point we do not have:

- tools
- memory
- external actions
- agent logic

```python
question = input("Ask a question: ")

response = client.responses.create(
    model="gpt-5",
    input=question
)

print(response.output_text)
```

## Key Takeaway

An LLM can answer questions and generate text.

An AI Agent is something more:

LLM + Tools + Memory + Logic

```python
response = client.responses.create(
    model="gpt-5",
    input="""
What is the current weather in Sofia right now?
"""
)

print(response.output_text)
```

## Why LLMs Are Not Enough

An LLM alone cannot:

- Access APIs
- Query databases
- Execute scripts
- Read Jira tickets
- Interact with external systems

To do that we need an Agent.
